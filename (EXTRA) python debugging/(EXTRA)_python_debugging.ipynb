## 📖 Overview

**`pdb`** (Python Debugger) is Python's built-in interactive debugger. It allows you to pause a running program at any point, inspect the current state of variables, step through code line by line, and identify exactly where and why a bug occurs — without adding `print()` statements everywhere.

<div style="background-color: #ebf8ff; border-left: 5px solid #3182ce; padding: 14px 18px; border-radius: 6px; margin-top: 16px;">
  <strong style="color: #2b6cb0;">🔑 Key Concept — Interactive Debugger:</strong><br/>
  <span style="color: #2d3748;">Unlike a regular Python shell, the <code>pdb</code> prompt appears <em>inside</em> a running program, giving you direct access to its execution context. You can interrogate variables, evaluate expressions, and control exactly how far the program advances.</span>
</div>

---
## Part 1 — Launching the Debugger

To start the debugger on a script, use the `-m` command-line flag to import the `pdb` module and pass your script as the argument:

```bash
python -m pdb looping.py
```

The `-m` flag tells Python to import any module you specify as a runnable program — in this case, `pdb`. Once executed, the debugger starts and displays its prompt:

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.95em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.8;">
  <span style="color: #68d391;">></span> /Users/sammy/looping.py(1)&lt;module&gt;()<br/>
  <span style="color: #f6e05e;">-></span> num_list = [500, 600, 700]<br/>
  <span style="color: #63b3ed;">(Pdb)</span> <span style="color: #fc8181;">|  &lt;&lt;== this is the pointer (current line)</span>
</div>

<div style="background-color: #fffff0; border-left: 5px solid #d69e2e; padding: 14px 18px; border-radius: 6px; margin-top: 4px;">
  <strong style="color: #b7791f;">💡 Notes:</strong><br/>
  <ul style="color: #2d3748; margin: 6px 0 0 0; padding-left: 20px;">
    <li>The <code>pdb</code> prompt (<code>Pdb)</code>) is distinct from the standard Python interactive shell.</li>
    <li>When the program reaches its end, the debugger <strong>restarts automatically</strong> from the beginning.</li>
    <li>Type <code>quit</code> or <code>exit</code> at any time to leave the debugger entirely.</li>
    <li>Type <code>run</code> to restart the debugger from the very first line of the program.</li>
  </ul>
</div>

## Part 2 — The Sample Program

Throughout this guide we will debug the following script, which contains two lists and a nested loop function. Save it as `looping.py` before running the debugger examples.

````python
# looping.py — the program we will debug

number_list = [500, 600, 700]
letter_list = ['x', 'y', 'z']


def nested_loop():
    """Iterates over number_list, and for each number iterates over letter_list."""
    for number in number_list:
        print(number)
        for letter in letter_list:
            print(letter)


if __name__ == '__main__':
    nested_loop()
````

## Part 3 — The `list` Command: Viewing Surrounding Code

Inside the `pdb` prompt, the `list` command displays the source code surrounding the **current execution line**, which is marked with `->`. This gives you visual context about where in the program the debugger is paused.

### 3.1 — `list` with no arguments (shows 11 surrounding lines)

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.92em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.9;">
  <span style="color: #63b3ed;">(Pdb)</span> list<br/>
  &nbsp; 1 <span style="color: #f6e05e;">-></span> num_list = [500, 600, 700]<br/>
  &nbsp; 2 &nbsp;&nbsp; alpha_list = ['x', 'y', 'z']<br/>
  &nbsp; 3<br/>
  &nbsp; 4<br/>
  &nbsp; 5 &nbsp;&nbsp; def nested_loop():<br/>
  &nbsp; 6 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; for number in num_list:<br/>
  &nbsp; 7 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; print(number)<br/>
  &nbsp; 8 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; for letter in alpha_list:<br/>
  &nbsp; 9 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; print(letter)<br/>
  10<br/>
  11 &nbsp;&nbsp; if __name__ == '__main__':<br/>
  <span style="color: #63b3ed;">(Pdb)</span>
</div>

The `->` arrow on line 1 indicates the **current execution point**. Because this is a short program, `list` shows the entire file. By default it shows the 11 lines most closely surrounding the current line.

### 3.2 — `list` with a line range argument

You can also specify which lines to display by passing a start and end line number:

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.92em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.9;">
  <span style="color: #63b3ed;">(Pdb)</span> list 3, 7<br/>
  &nbsp; 3<br/>
  &nbsp; 4<br/>
  &nbsp; 5 &nbsp;&nbsp; def nested_loop():<br/>
  &nbsp; 6 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; for number in num_list:<br/>
  &nbsp; 7 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; print(number)<br/>
  <span style="color: #63b3ed;">(Pdb)</span>
</div>

## Part 4 — Stepping Through Code: `step` vs `next`

Both `step` and `next` advance execution by one line, but they behave differently when encountering a **function call**.

<table style="width:100%; border-collapse: collapse; font-family: 'Segoe UI', sans-serif; font-size: 0.94em; margin-top: 8px;">
  <thead>
    <tr style="background-color: #2d3748; color: #e2e8f0;">
      <th style="padding: 12px 16px; text-align: left;">Command</th>
      <th style="padding: 12px 16px; text-align: left;">Shortcut</th>
      <th style="padding: 12px 16px; text-align: left;">Behaviour at a Function Call</th>
    </tr>
  </thead>
  <tbody>
    <tr style="background-color: #edf2f7;">
      <td style="padding: 11px 16px; font-family: monospace;">step</td>
      <td style="padding: 11px 16px; font-family: monospace;">s</td>
      <td style="padding: 11px 16px;"><strong>Steps <em>into</em></strong> the called function, pausing at its first line. Useful for inspecting what the function does internally.</td>
    </tr>
    <tr style="background-color: #ffffff;">
      <td style="padding: 11px 16px; font-family: monospace;">next</td>
      <td style="padding: 11px 16px; font-family: monospace;">n</td>
      <td style="padding: 11px 16px;"><strong>Steps <em>over</em></strong> the called function, executing it in full and pausing at the next line in the current scope. Useful when you trust the function and only care about its result.</td>
    </tr>
  </tbody>
</table>

<div style="background-color: #f0fff4; border-left: 5px solid #38a169; padding: 14px 18px; border-radius: 6px; margin-top: 14px;">
  <strong style="color: #276749;">🔍 Practical Example:</strong><br/>
  <span style="color: #2d3748;">When using <code>step</code> on our <code>nested_loop()</code> function, the debugger will walk through every iteration of both loops — printing each <code>number</code>, then each <code>letter</code>, then looping back — revealing the exact interleaved execution order. Using <code>next</code> instead would execute the entire function at once and pause after it returns.</span>
</div>

<div style="background-color: #ebf8ff; border-left: 5px solid #3182ce; padding: 14px 18px; border-radius: 6px; margin-top: 10px;">
  <strong style="color: #2b6cb0;">⌨️ Tip — Repeat Last Command:</strong><br/>
  <span style="color: #2d3748;">Pressing <kbd>Enter</kbd> without typing anything at the <code>(Pdb)</code> prompt repeats the most recently executed debugger command — a time-saver when stepping through many lines.</span>
</div>

## Part 5 — Inspecting Variables: The `pp` Command

While stepping through code, you will often want to inspect the current value held by a variable. The `pp` command (**pretty-print**) evaluates any expression you pass to it and displays its value using Python's `pprint` module for clean, readable formatting.

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.92em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.9;">
  <span style="color: #63b3ed;">(Pdb)</span> pp number_list<br/>
  [500, 600, 700]<br/>
  <span style="color: #63b3ed;">(Pdb)</span> pp letter_list<br/>
  ['x', 'y', 'z']<br/>
  <span style="color: #63b3ed;">(Pdb)</span>
</div>

You can also evaluate expressions directly:

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.92em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.9;">
  <span style="color: #63b3ed;">(Pdb)</span> pp len(number_list)<br/>
  3
</div>

<div style="background-color: #fff5f5; border-left: 5px solid #e53e3e; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #c53030;">📝 Note — Most <code>pdb</code> commands have shorthand aliases:</strong><br/>
  <span style="color: #2d3748;">For example, <code>step</code> → <code>s</code>, <code>next</code> → <code>n</code>, <code>list</code> → <code>l</code>, <code>break</code> → <code>b</code>. Type <code>help</code> at the prompt for the full list of available commands and their abbreviations.</span>
</div>

## Part 6 — Breakpoints: Jumping to a Specific Line

Stepping through a large program line by line is impractical. **Breakpoints** let you instruct the debugger to run freely until it reaches a specific line, then pause and hand control back to you.

Set a breakpoint using the `break` command followed by the filename and line number:

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.92em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.9;">
  <span style="color: #63b3ed;">(Pdb)</span> break looping.py:5<br/>
  Breakpoint 1 at /Users/sammy/looping.py:5<br/>
  <span style="color: #63b3ed;">(Pdb)</span>
</div>

The debugger assigns each breakpoint a **sequential number** starting from `1`. These numbers are used to manage breakpoints (e.g., enabling, disabling, or removing individual ones).

<div style="background-color: #fffff0; border-left: 5px solid #d69e2e; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #b7791f;">💡 Breakpoint Syntax:</strong><br/>
  <code style="color: #2d3748;">(Pdb) break &lt;filename&gt;:&lt;line_number&gt;</code><br/><br/>
  <span style="color: #2d3748;">You may also set breakpoints on a function name directly:<br/>
  <code>(Pdb) break looping.nested_loop</code></span>
</div>

To **remove all breakpoints** at once, type `clear` and confirm with `y` when prompted:

<div style="background-color: #1a202c; color: #a0aec0; font-family: monospace; font-size: 0.92em; padding: 16px 22px; border-radius: 8px; margin: 14px 0; line-height: 1.9;">
  <span style="color: #63b3ed;">(Pdb)</span> clear<br/>
  Clear all breaks? y<br/>
  Deleted breakpoint 1 at /Users/sammy/looping.py:5<br/>
  <span style="color: #63b3ed;">(Pdb)</span>
</div>

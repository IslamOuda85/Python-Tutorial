## 📦 Step 1 — Importing Required Modules

We import the built-in `random` module, which provides utilities for generating pseudo-random numbers. This will be used later to simulate a coin toss.

````python
import random  # Provides random number generation utilities
````

## 🏗️ Step 2 — Defining the `Coin` Class

<div style="background:#fdfefe; border:1px solid #d5d8dc; border-radius:8px; padding:16px; font-family:sans-serif;">

<p>A <strong>class</strong> is a blueprint for creating objects. By convention, class names are written in
<strong>PascalCase</strong> (e.g. <code>Coin</code>, not <code>coin</code>) to distinguish them clearly from regular variable names.</p>

<p>The <code>Coin</code> class models a physical coin with two possible states: <em>heads</em> or <em>tails</em>.</p>
</div>

### 🔧 The `__init__` Method — Object Initializer

<div style="background:#eafaf1; border-left:5px solid #27ae60; padding:12px 18px; border-radius:4px; font-family:sans-serif;">
  <p><code>__init__</code> is a <strong>special (dunder) method</strong> that Python calls <em>automatically</em> the moment a new object is created from the class.</p>
  <p>Its primary responsibility is to <strong>initialize the object's attributes</strong> — setting up the internal state that each object will carry from birth.</p>
</div>

```
def __init__(self):
    self.sideup = 'heads'
    │    │              │
    │    │              └─ The default starting value of the attribute
    │    └──────────────── The name of the instance attribute being created
    └───────────────────── 'self' refers to the object being initialized
```

> **Key point:** `self.sideup = 'heads'` creates an **instance attribute** named `sideup` and assigns it the default value `'heads'` whenever a new `Coin` object is created.

### 🪪 The `self` Parameter — The Object's Identity

<div style="background:#fef9e7; border-left:5px solid #f39c12; padding:12px 18px; border-radius:4px; font-family:sans-serif;">

<p>Because a class can produce <strong>many independent objects</strong>, Python needs a way to know <em>which specific object</em> a method is operating on. That is the role of <code>self</code>.</p>

<ul>
  <li><code>self</code> is always the <strong>first parameter</strong> of every instance method.</li>
  <li>When you call <code>coin_instance.toss()</code>, Python automatically passes the object <code>coin_instance</code> as the <code>self</code> argument — you never pass it manually.</li>
  <li>Using <code>self</code>, all methods inside a class share and modify the <strong>same object's state</strong>, unlike ordinary functions that have isolated local scopes.</li>
</ul>

</div>

| Context | Behaviour |
|---|---|
| Two independent functions | Cannot access each other's local variables |
| Two methods inside a class | Share state freely through `self.attribute` |

````python
class Coin:
    """
    Represents a coin that can be tossed to land on either 'heads' or 'tails'.

    Attributes
    ----------
    sideup : str
        The current face of the coin showing upward. Defaults to 'heads'.
    """

    def __init__(self):
        """
        Initializes a new Coin object.
        Sets the default face-up side to 'heads'.
        This method is called automatically upon object creation.
        """
        self.sideup = 'heads'

    def toss(self):
        """
        Simulates a coin toss by randomly setting the face-up side
        to either 'heads' or 'tails' with equal probability.
        """
        if random.randint(0, 1) == 0:
            self.sideup = 'heads'
        else:
            self.sideup = 'tails'

    def get_sideup(self):
        """
        Getter method — returns the current face-up side of the coin.
        Provides controlled, read-only access to the sideup attribute.

        Returns
        -------
        str
            Either 'heads' or 'tails'.
        """
        return self.sideup
````

## 🪙 Step 3 — Creating and Using an Object

### Object Instantiation Formula

```python
# Local class
variable = ClassName()

# Imported class
variable = module_name.ClassName()
```

Calling `Coin()` triggers `__init__` automatically, so `coin_instance.sideup` is already set to `'heads'` before we do anything else.

````python
# Create a new Coin object — __init__ runs immediately, setting sideup = 'heads'
coin_instance = Coin()

# Access the initial face-up side via the getter method
print("Initial side up  :", coin_instance.get_sideup())   # Expected: 'heads'

# Toss the coin — randomly updates sideup; the method itself returns None
print("Return of toss() :", coin_instance.toss())          # Expected: None

# Read the updated face-up side directly from the attribute
print("After toss       :", coin_instance.sideup)          # Expected: 'heads' or 'tails'
````

<div style="background:#f4f6f7; border:1px solid #abb2b9; border-radius:6px; padding:14px 18px; font-family:sans-serif;">

**📌 Why does `toss()` take no arguments when called, yet its definition has `self`?**

Python automatically injects the calling object as the first argument.  
So `coin_instance.toss()` is silently translated to `Coin.toss(coin_instance)` behind the scenes.

</div>

## 🔬 Step 4 — Why `self` Matters: Scope Comparison

The following example demonstrates the fundamental difference between **ordinary functions** and **class methods**:
ordinary functions cannot share local variables with one another, whereas class methods are connected through `self`.

````python
# --- Ordinary functions: isolated local scopes ---

def function_a():
    local_value = 123

def function_b():
    print(local_value)  # NameError: 'local_value' is not accessible here

function_a()
# function_b()  # <-- Uncommenting this line raises a NameError

print("Ordinary functions cannot share local variables with each other.")
print("Class methods CAN share state through 'self.attribute'.")
````

## ⚠️ Step 5 — The Danger of Unprotected Attributes

<div style="background:#fdedec; border-left:5px solid #e74c3c; padding:12px 18px; border-radius:4px; font-family:sans-serif;">

<strong>Security Concern:</strong> Because <code>sideup</code> is a public attribute, any external code can overwrite it with <em>any</em> value — including nonsensical ones. This is why <strong>attribute hiding (encapsulation)</strong> is introduced in the next example.

</div>

The cell below demonstrates how an unprotected public attribute can be corrupted:

````python
# Create a second independent Coin object
vulnerable_coin = Coin()

# Valid assignment — but bypasses any validation logic
vulnerable_coin.sideup = 'tails'

# Invalid assignments — illustrate the lack of data protection
vulnerable_coin.sideup = 'batata'   # Not a valid coin face
vulnerable_coin.sideup = 0.0        # Wrong data type entirely
vulnerable_coin.sideup = print('⚠️  WARNING: Attribute overwritten with a function call result!')

# Toss the coin and display its (now meaningless) state
vulnerable_coin.toss()
print("Final side up on vulnerable_coin:", vulnerable_coin.get_sideup())
````

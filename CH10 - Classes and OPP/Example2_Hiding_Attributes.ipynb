## 📖 Overview

In the previous example, all class attributes and methods were **public** — freely readable and modifiable from outside the class. While convenient, this openness can lead to accidental misuse or unintended modifications.

This notebook explores **access restriction** in Python classes: how to signal that certain attributes and methods are intended to be *private*, and what that truly means in practice.

<div style="background-color: #ebf8ff; border-left: 5px solid #3182ce; padding: 14px 18px; border-radius: 6px; margin-top: 16px;">
  <strong style="color: #2b6cb0;">🔑 Key Concept — Name Mangling:</strong><br/>
  <span style="color: #2d3748;">Python does <em>not</em> enforce true private access. Instead, it uses a convention called <strong>name mangling</strong>: any attribute or method prefixed with two underscores (<code>__</code>) is automatically renamed internally to <code>_ClassName__attributeName</code>, making accidental access harder — but not impossible.</span>
</div>

---
## Part 1 — Declaring Private Members with `__` (Double Underscore)

To mark an attribute or method as **private**, prefix its name with two underscores (`__`). This signals to other developers (and tools) that it is an **internal implementation detail** and should not be accessed directly from outside the class.

<div style="background-color: #fff5f5; border-left: 5px solid #e53e3e; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #c53030;">⚠️ Attempting to access private members directly will raise an <code>AttributeError</code>.</strong>
</div>

```python
class SecureContainer:
    """A class demonstrating private attributes and methods via name mangling."""

    def __init__(self):
        self.__confidential_value = 10  # Private instance attribute

    def __confidential_method(self):    # Private method
        return 10


secure_instance = SecureContainer()

# Attempting direct access — both lines below raise AttributeError:
# print(secure_instance.__confidential_value)    # ERROR
# print(secure_instance.__confidential_method()) # ERROR

print("Direct access to private members is blocked — as expected.")
```

---
## Part 2 — Python's Name Mangling Loophole

Python does **not** truly enforce privacy. The double-underscore prefix triggers **name mangling**, which internally renames the member to `_ClassName__memberName`. This means it can still be accessed if you know the mangled name.

<div style="background-color: #fffff0; border-left: 5px solid #d69e2e; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #b7791f;">💡 This is a known Python design quirk — not a security feature.</strong><br/>
  <span style="color: #2d3748;">The double-underscore convention exists to <em>discourage</em> accidental access, not to enforce it. Never rely on it for true data security.</span>
</div>

```python
# Accessing private members via their mangled names (use with caution!)
print(secure_instance._SecureContainer__confidential_method())  # Output: 10
print(secure_instance._SecureContainer__confidential_value)     # Output: 10
```

---
## Part 3 — The Proper Solution: Getter Methods

The **recommended and idiomatic** approach to controlled access is to expose private data through a **public getter method**. This allows the class to:

- Control *what* is returned (e.g., a copy, a formatted value).
- Add validation or logging before returning data.
- Keep the internal attribute truly hidden from direct manipulation.

<div style="background-color: #f0fff4; border-left: 5px solid #38a169; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #276749;">✅ Best Practice:</strong><br/>
  <span style="color: #2d3748;">Always access private attributes through <strong>getter methods</strong> (or Python's built-in <code>@property</code> decorator). Never rely on the name mangling workaround shown in Part 2.</span>
</div>

```python
class ProtectedRecord:
    """Demonstrates controlled access to a private attribute via a getter method."""

    __restricted_value = 100  # Private class-level attribute

    def get_restricted_value(self):
        """Public getter method — the sanctioned way to read the private attribute."""
        return self.__restricted_value


record_instance = ProtectedRecord()

# Direct access is blocked:
# print(record_instance.__restricted_value)  # ERROR

# Correct approach — use the getter:
print(record_instance.get_restricted_value())  # Output: 100
```

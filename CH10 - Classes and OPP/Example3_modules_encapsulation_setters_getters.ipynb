## 📁 Part 1 — Organising Classes into Separate Modules

### Why Use Separate Modules?

For small, exploratory scripts it is perfectly acceptable to define classes and the code that uses them in a single file. However, as a program grows, keeping everything in one file becomes difficult to navigate and maintain.

<div style="background:#f4f6f7; border:1px solid #abb2b9; border-radius:8px; padding:14px 20px; font-family:sans-serif;">

| Approach | Suitable for | Drawbacks when scaling |
|---|---|---|
| Single file (class + logic) | Quick scripts, prototypes | Hard to reuse, navigate, or debug |
| Separate module per class | Medium–large projects | None — this is the recommended practice |

</div>

By placing the `Coin` class inside its own file (`classes.py`), we gain three immediate benefits:

1. **Reusability** — any other script can import the class without copy-pasting code.
2. **Readability** — each file has a single, well-defined responsibility.
3. **Maintainability** — bugs and improvements are isolated to one location.

## 🔒 Part 2 — Encapsulation: Hiding Attributes with `__` (Name Mangling)

In the previous example, `sideup` was a **public** attribute — any code outside the class could overwrite it with an invalid value (e.g. a number or a nonsense string).

Python's solution is **name mangling**: prefix an attribute name with two underscores (`__`) to make it private to the class.

<div style="background:#eafaf1; border-left:5px solid #27ae60; padding:12px 18px; border-radius:4px; font-family:sans-serif;">

| Attribute syntax | Visibility | Accessible from outside the class? |
|---|---|---|
| `self.sideup` | Public | ✅ Yes — anyone can read or overwrite it |
| `self.__sideup` | Private (name-mangled) | ❌ No — direct access raises an `AttributeError` |

</div>

When Python encounters `self.__sideup`, it internally renames the attribute to `_Coin__sideup`, making it inaccessible under its original name from outside the class — this mechanism is called **name mangling**.

## 🏗️ Part 3 — The `Coin` Class: Full Definition

Below is the complete `Coin` class as it would appear in `classes.py`. It now features:

- A **private** `__sideup` attribute
- A `flip()` method that toggles the current face
- A `get_sideup()` **getter** for read access
- A `set_sideup()` **setter** for validated write access

```python
import random


class Coin:
    """
    Represents a coin with a hidden (private) face-up attribute.

    The internal state is fully encapsulated: outside code cannot read or
    modify the face-up side directly. All access must go through the
    provided getter and setter methods.

    Attributes
    ----------
    __sideup : str
        The current face showing upward. Private — accessible only within
        this class. Valid values: 'heads', 'tails'.
    """

    def __init__(self):
        """
        Initialises the coin with 'heads' facing upward.
        The double-underscore prefix makes __sideup a private attribute,
        preventing direct external access or modification.
        """
        self.__sideup = 'heads'

    def flip(self):
        """
        Toggles the coin's face-up side.
        If the current side is 'heads', it becomes 'tails', and vice versa.
        This simulates a physical coin flip with guaranteed state alternation.
        """
        if self.__sideup == 'heads':
            self.__sideup = 'tails'
        else:
            self.__sideup = 'heads'

    def get_sideup(self):
        """
        Getter — returns the current face-up side of the coin.

        Returns
        -------
        str
            Either 'heads' or 'tails'.
        """
        return self.__sideup

    def set_sideup(self, selection: str):
        """
        Setter — updates the face-up side based on a validated user selection.

        Parameters
        ----------
        selection : str
            '1' to set the face to 'heads'; '2' to set it to 'tails'.
            Any other value is silently ignored, preserving the current state.
        """
        if selection == '1':
            self.__sideup = 'heads'
        elif selection == '2':
            self.__sideup = 'tails'
```

## 🔑 Part 4 — Getters and Setters

Because the `__sideup` attribute is private, outside code cannot read or modify it directly. **Getter** and **setter** methods serve as the controlled doorways into the class's internal state.

### General Patterns

<div style="display:flex; gap:20px; font-family:monospace; font-size:13px;">

<div style="flex:1; background:#eafaf1; border:1px solid #a9dfbf; border-radius:6px; padding:14px;">
<strong style="font-family:sans-serif; color:#1e8449;">✅ Getter</strong><br><br>
<pre style="margin:0;">def get_attribute_name(self):
    return self.__attribute_name</pre>
</div>

<div style="flex:1; background:#eaf4fb; border:1px solid #a9cce3; border-radius:6px; padding:14px;">
<strong style="font-family:sans-serif; color:#1a5276;">✅ Setter</strong><br><br>
<pre style="margin:0;">def set_attribute_name(self, new_value):
    self.__attribute_name = new_value</pre>
</div>

</div>

<br>

<div style="background:#fef9e7; border-left:5px solid #f39c12; padding:12px 18px; border-radius:4px; font-family:sans-serif;">
  <strong>📐 Design Rule:</strong><br>
  In a well-designed class, the number of getters, setters, and private data attributes should all be equal —
  one getter and one setter per private attribute.

  <br><br>

  <code>count(getters) = count(setters) = count(private data attributes)</code>

  <br><br>

  In our <code>Coin</code> class: <strong>1 private attribute</strong> (<code>__sideup</code>)
  → <strong>1 getter</strong> (<code>get_sideup</code>) → <strong>1 setter</strong> (<code>set_sideup</code>).
</div>

### 🔬 Demonstrating Encapsulation: Private vs. Public Access

```python
coin_instance = Coin()

# ✅ Correct — using the getter to read the private attribute
print("Via getter             :", coin_instance.get_sideup())  # 'heads'

# ❌ Incorrect — attempting to access the private attribute directly
try:
    print(coin_instance.__sideup)
except AttributeError as error:
    print("Direct access blocked  :", error)

# ✅ Correct — using the setter to update the private attribute with validation
coin_instance.set_sideup('2')
print("After set_sideup('2') :", coin_instance.get_sideup())  # 'tails'

# ✅ Correct — invalid input is silently ignored; state is preserved
coin_instance.set_sideup('invalid')
print("After invalid input    :", coin_instance.get_sideup())  # 'tails' (unchanged)
```

## ▶️ Part 5 — Driving Code (Equivalent of `main.py`)

The following cell replicates the logic from `main.py`. It demonstrates how to use an imported `Coin` object: creating an instance, reading its state, and flipping it multiple times.

> In a real multi-file project this code would live in `main.py` and begin with `import classes`. Here, since the class is already defined above in this notebook, we use it directly.

```python
# Equivalent of: import classes
# Equivalent of: coin_instance = classes.Coin()
coin_instance = Coin()

print("Initial state :", coin_instance.get_sideup())   # Expected: heads

coin_instance.flip()
print("After flip 1  :", coin_instance.get_sideup())   # Expected: tails

coin_instance.flip()
print("After flip 2  :", coin_instance.get_sideup())   # Expected: heads

coin_instance.flip()
print("After flip 3  :", coin_instance.get_sideup())   # Expected: tails
```

<div style="background:#f4f6f7; border:1px solid #abb2b9; border-radius:6px; padding:14px 18px; font-family:sans-serif;">

**📌 `flip()` vs. `toss()` — What's the difference?**

| Method | Behaviour | Deterministic? |
|---|---|---|
| `flip()` | Always switches to the opposite face | ✅ Yes — predictable, always alternates |
| `toss()` *(Example 1)* | Randomly selects heads or tails | ❌ No — outcome is random each call |

Choose `flip()` when you need guaranteed alternation; choose `toss()` when you need true randomness.
</div>

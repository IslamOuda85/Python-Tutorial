## 📖 Overview

Have you ever tried to `print()` an object directly and seen something like this?

<div style="background-color: #1a1a2e; color: #e53e3e; font-family: monospace; font-size: 1em; padding: 14px 20px; border-radius: 8px; margin: 12px 0;">
  &lt;__main__.ClassName object at 0x0000020814E0A570&gt;
</div>

That cryptic output is Python telling you the **type** and **memory address** of the object — useful for debugging, but meaningless to an end user.

In this notebook, we learn how to override this default behavior using the **`__str__` magic method**, which lets you define a clean, human-readable string representation for any class you create.

<div style="background-color: #ebf8ff; border-left: 5px solid #3182ce; padding: 14px 18px; border-radius: 6px; margin-top: 16px;">
  <strong style="color: #2b6cb0;">🔑 Key Concept — Magic Methods (Dunder Methods):</strong><br/>
  <span style="color: #2d3748;">Methods surrounded by double underscores (e.g. <code>__str__</code>, <code>__init__</code>) are called <strong>dunder</strong> (double-underscore) or <strong>magic</strong> methods. Python calls them automatically in response to built-in operations — you rarely call them directly.</span>
</div>

---
## Part 1 — Without `__str__`: The Default Memory-Address Output

The class below represents a starship and stores four statistics. It exposes those stats through a regular method called `get_stats()`. Notice what happens when you `print()` the object itself versus when you call `get_stats()` explicitly.

<div style="background-color: #fff5f5; border-left: 5px solid #e53e3e; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #c53030;">⚠️ Without <code>__str__</code>:</strong> <code>print(object)</code> outputs a raw memory reference — not the data you care about. You must call a custom method manually every time.
</div>

```python
class StarshipWithoutStr:
    """A starship class that lacks a __str__ method.
    The caller must invoke get_stats() explicitly to see meaningful output.
    """

    def __init__(self):
        self.name   = 'Orion'
        self.damage = 663
        self.speed  = 8.33
        self.health = 619

    def get_stats(self):
        """Returns a formatted summary of the starship's statistics."""
        return (
            f'Name:   {self.name}\n'
            f'Damage: {self.damage}\n'
            f'Speed:  {self.speed}\n'
            f'Health: {self.health}'
        )


ship_without_str = StarshipWithoutStr()

print("--- print(object) ---")
print(ship_without_str)           # Outputs raw memory address — not useful

print("\n--- print(object.get_stats()) ---")
print(ship_without_str.get_stats())  # Must call the method explicitly
```

---
## Part 2 — With `__str__`: Clean, Automatic String Representation

By defining a `__str__` method, you instruct Python on **exactly what to display** whenever the object is converted to a string — whether via `print()`, `str()`, or an f-string.

<div style="background-color: #f0fff4; border-left: 5px solid #38a169; padding: 14px 18px; border-radius: 6px;">
  <strong style="color: #276749;">✅ With <code>__str__</code>:</strong> <code>print(object)</code> automatically calls <code>__str__</code> and displays your formatted output — no extra method call required.
</div>

```python
class StarshipWithStr:
    """A starship class that implements __str__ for clean string output.
    Any built-in string conversion of this object will use __str__ automatically.
    """

    def __init__(self):
        self.name   = 'Orion'
        self.damage = 663
        self.speed  = 8.33
        self.health = 619

    def __str__(self):
        """Returns a human-readable summary of the starship's statistics.
        Called automatically by print(), str(), and f-strings.
        """
        return (
            f'Name:   {self.name}\n'
            f'Damage: {self.damage}\n'
            f'Speed:  {self.speed}\n'
            f'Health: {self.health}'
        )


ship_with_str = StarshipWithStr()

print("--- print(object) ---")
print(ship_with_str)              # __str__ is called automatically

print("\n--- object.__str__() called directly ---")
print(ship_with_str.__str__())    # Explicit call — same result

print("\n--- str(object) ---")
print(str(ship_with_str))         # str() also delegates to __str__

print("\n--- str() on a freshly constructed instance ---")
print(str(StarshipWithStr()))     # Works on any new instance
```

## 📝 Summary

<div style="background-color: #fffff0; border-left: 5px solid #d69e2e; padding: 14px 18px; border-radius: 6px; margin-bottom: 16px;">
  <strong style="color: #b7791f;">💡 Rule of Thumb:</strong><br/>
  <span style="color: #2d3748;">Define <code>__str__</code> in every class that represents a meaningful entity (a user, a product, a game character, etc.). It costs one method and pays dividends every time the object is printed, logged, or embedded in a larger string.</span>
</div>

| Method | Triggered by | Purpose |
|---|---|---|
| `__init__` | `ClassName()` | Initialise instance attributes |
| `__str__` | `print()`, `str()`, f-strings | Human-readable string representation |
| `get_stats()` *(custom)* | Explicit call only | Works, but requires caller knowledge |

# 🗂️ PART 1 — WORKING WITH FILES

Files let your program **save data permanently** — even after the program closes, the data lives on disk. Python makes working with files simple through the built-in `open()` function.

## 1. Opening a File — `open()` & File Modes

The `open()` function connects your program to a file on disk:

```python
file_object = open('filename.txt', 'mode')
```

### File Modes

| Mode | Name | Behaviour | File exists? | File missing? |
|---|---|---|---|---|
| `'r'` | Read | Read only, starts at beginning | Opens it | ❌ `FileNotFoundError` |
| `'w'` | Write | **Overwrites** entire file | Clears it | Creates it |
| `'a'` | Append | Adds to the **end** of the file | Keeps old data | Creates it |
| `'r+'` | Read+Write | Read and write without overwriting | Opens it | ❌ `FileNotFoundError` |
| `'rb'` / `'wb'` | Binary | Read/write raw binary data (e.g. images, pickle) | — | — |

> ⚠️ **Key rule:** Opening in `'w'` mode **immediately erases** all existing content. Use `'a'` if you want to keep what's already there.

````python
# Just opening a file — we'll write and read in the next sections
# This shows that open() returns a file object

f = open('demo.txt', 'w')   # creates (or clears) demo.txt
print('File object :', f)
print('File name   :', f.name)
print('File mode   :', f.mode)
f.close()   # always close after opening!
````

## 2. Writing to a File — `write()`

Once a file is open in `'w'` or `'a'` mode, use `.write()` to add text:

```python
file.write('some text')          # writes a string
file.write(f'value: {variable}') # f-strings work here too
```

> ⚠️ Python does **not** add newlines automatically. You must include `\n` yourself wherever you want a line break.

````python
name   = 'Islam'
origin = 'Gaza'

filey = open('my_file.txt', 'w')              # 'w' = write (overwrites)
filey.write(f'My name is {name}\n')           # \n = newline
filey.write(f'I am from {origin}\n')
filey.write('I love pizza')
filey.close()

print('File written successfully.')

# Verify by reading it back right away
filey = open('my_file.txt', 'r')
print(filey.read())
filey.close()
````

## 3. Reading a File — `readline()`, `readlines()`, `read()`

There are three ways to read file content:

| Method | Returns | Reads |
|---|---|---|
| `file.read()` | One big `str` | The **entire** file at once |
| `file.readline()` | One `str` | One line per call (includes `\n`) |
| `file.readlines()` | `list` of `str` | All lines as a list (each includes `\n`) |

> 📌 **Remember:** Each call to `readline()` advances an internal cursor. Calling it again reads the **next** line.

````python
# --- read() — entire file as one string ---
filey = open('my_file.txt', 'r')
content = filey.read()
filey.close()

print('=== read() ===')
print(content)
print('Type:', type(content))
````

````python
# --- readline() — one line at a time ---
filey = open('my_file.txt', 'r')

line1 = filey.readline()   # reads first line
line2 = filey.readline()   # reads second line
line3 = filey.readline()   # reads third line

filey.close()

print('=== readline() ===')
print(repr(line1))   # repr() shows the \n clearly
print(repr(line2))
print(repr(line3))
````

````python
# --- readlines() — all lines as a list ---
filey = open('my_file.txt', 'r')
lines = filey.readlines()
filey.close()

print('=== readlines() ===')
print('Type    :', type(lines))
print('Content :', lines)   # each element includes \n
````

## 4. Cleaning Lines — `rstrip()` & `strip()`

When reading lines from a file, each line ends with a `\n` character. Use string methods to remove it:

| Method | Removes from | What it removes |
|---|---|---|
| `str.rstrip('\n')` | Right end only | The specified character(s) |
| `str.strip()` | Both ends | All whitespace (`\n`, `\t`, spaces) |
| `str.lstrip()` | Left end only | All whitespace |


````python
filey = open('my_file.txt', 'r')
raw_line = filey.readline()   # 'My name is Islam\n'
filey.close()

print('Raw line (with \\n) :', repr(raw_line))

# rstrip removes from the RIGHT end
clean_rstrip = raw_line.rstrip('\n')
print('After rstrip("\\n") :', repr(clean_rstrip))

# strip removes from BOTH ends (whitespace by default)
clean_strip = raw_line.strip()
print('After strip()      :', repr(clean_strip))

# Practical: clean all lines in a list
filey = open('my_file.txt', 'r')
cleaned_lines = [line.strip() for line in filey.readlines()]
filey.close()
print('\nAll lines cleaned  :', cleaned_lines)
````

## 5. Closing Files — `close()` & the `with` Statement

### Why close a file?
- Flushes any unwritten data from memory to disk
- Releases the file lock so other programs can access it
- Especially critical for large files or long-running programs

### The problem with manual `close()`
If an error occurs **between** `open()` and `close()`, the file never gets closed.

### ✅ The solution: `with` statement (context manager)
The `with` statement **automatically closes the file** when the block ends — even if an exception occurs.

```python
# ❌ Risky — if an error occurs, close() is never called
f = open('file.txt', 'r')
data = f.read()
f.close()

# ✅ Safe — file is always closed automatically
with open('file.txt', 'r') as f:
    data = f.read()
# file is closed here, automatically
```

````python
# Writing with 'with'
with open('my_file.txt', 'w') as f:
    f.write('My name is Islam\n')
    f.write('I am from Gaza\n')
    f.write('I love pizza')
# File is automatically closed here

print('Is file closed after with block?', f.closed)   # True

# Reading with 'with'
with open('my_file.txt', 'r') as f:
    content = f.read()

print('\nFile contents:')
print(content)
````

## 6. Appending to a File

Opening in `'a'` (append) mode adds new content to the **end** of the file — it never erases existing data.

| Mode | What happens to existing content |
|---|---|
| `'w'` | **Erased** immediately on open |
| `'a'` | **Preserved** — new content goes after it |

````python
# Append a new line to the existing file
with open('my_file.txt', 'a') as f:
    f.write('\nAnd I enjoy coding!')

# Read back the full file to confirm
with open('my_file.txt', 'r') as f:
    print(f.read())
````

## 7. Looping Through a File Line by Line

The most memory-efficient way to read a file is to iterate over it directly in a `for` loop. Python reads **one line at a time** without loading the whole file into memory — ideal for large files.

````python
# Write a sample multi-line file
with open('scores.txt', 'w') as f:
    f.write('Alice: 95\nBob: 82\nCharlie: 78\nDiana: 91\n')

# Loop line by line (memory-efficient)
print('--- Line by line ---')
with open('scores.txt', 'r') as f:
    for line in f:
        print(line.strip())   # strip() removes trailing \n

# With line numbers using enumerate()
print('\n--- With line numbers ---')
with open('scores.txt', 'r') as f:
    for i, line in enumerate(f, start=1):
        print(f'  Line {i}: {line.strip()}')
````

# ⚠️ PART 2 — EXCEPTION HANDLING

An **exception** is a runtime error — something unexpected that happens while the program is running. Without handling exceptions, your program **crashes**. With handling, you can respond gracefully and keep running.

## 8. What is an Exception?

When Python encounters an error it cannot handle automatically, it **raises** an exception. Execution stops and an error message (a **traceback**) is shown.

```
Traceback (most recent call last):
  File "main.py", line 2, in <module>
    x = int('hello')
ValueError: invalid literal for int() with base 10: 'hello'
         ↑ Exception type      ↑ Description
```

````python
# Without handling — this would crash the program:
# x = int('hello')   # ValueError
# y = 1 / 0          # ZeroDivisionError
# z = [1,2,3][10]    # IndexError

# Instead, let's intentionally trigger errors and catch them:
examples = [
    ("int('hello')",    lambda: int('hello')),
    ("1 / 0",           lambda: 1 / 0),
    ("[1,2,3][10]",     lambda: [1,2,3][10]),
]

for desc, fn in examples:
    try:
        fn()
    except Exception as e:
        print(f'{type(e).__name__}: {e}   ←  triggered by: {desc}')
````

## 9. `try` / `except`

The basic structure:

```python
try:
    # Code that MIGHT raise an exception
    risky_code()
except ExceptionType:
    # Code that runs ONLY IF that exception occurs
    handle_the_error()
```

- The `try` block is **always attempted**
- If an exception occurs → Python jumps to `except` immediately
- Code after the error in `try` is **skipped**

````python
# Without try/except — would crash:
# result = int('abc')

# With try/except — handles it gracefully:
for test_input in ['42', 'abc', '7', '!@#']:
    try:
        result = int(test_input)
        print(f'  {test_input!r:6} → converted successfully: {result}')
    except ValueError:
        print(f'  {test_input!r:6} → cannot convert to integer!')
````

## 10. Catching Specific Exceptions

You can capture the **exception object** itself using `as` — this gives you access to the error message:

```python
except ValueError as e:
    print(e)   # prints the error description
```

> 💡 The variable name after `as` (often `e` or `err`) is just a convention — you can name it anything.

````python
test_values = ['hello', '3.14', '-5', 'NaN']

for val in test_values:
    try:
        number = int(val)
        print(f'Success: {val!r} → {number}')
    except ValueError as e:
        # 'e' holds the full error message from Python
        print(f'ValueError caught: {e}')
````

## 11. Multiple `except` Clauses

A single `try` block can have **multiple `except` clauses** — one for each type of error you want to handle differently.

```python
try:
    ...
except ValueError:
    ...   # handles bad type conversions
except ZeroDivisionError:
    ...   # handles division by zero
except FileNotFoundError:
    ...   # handles missing files
except Exception as e:
    ...   # catch-all for anything else
```

> ⚠️ **Order matters:** Put specific exceptions **before** generic ones. `Exception` catches everything, so it should always be last.

````python
def safe_divide(a, b):
    try:
        result = int(a) / int(b)
        return result
    except ValueError as e:
        print(f'  ValueError: {e}')
    except ZeroDivisionError:
        print(f'  ZeroDivisionError: cannot divide {a} by zero')
    except Exception as e:
        print(f'  Unexpected error ({type(e).__name__}): {e}')
    return None

tests = [('10', '2'), ('10', '0'), ('10', 'x'), (None, '2')]

for a, b in tests:
    print(f'safe_divide({a!r}, {b!r}):')
    result = safe_divide(a, b)
    if result is not None:
        print(f'  Result: {result}')
````

## 12. `else` — Runs Only When No Exception Occurs

The optional `else` block runs **only** if the `try` block completed **without raising any exception**.

```
try → exception raised   →  except runs,  else is SKIPPED
try → no exception       →  except skipped, else RUNS
```

> 💡 Use `else` for code that **should only run on success** — keeping it out of `try` makes the logic clearer and avoids accidentally catching errors from the success-path code.

````python
for test in ['42', 'hello']:
    print(f"--- Input: {test!r} ---")
    try:
        value = int(test)
    except ValueError as e:
        print(f'  EXCEPT : {e}')
    else:
        # Only runs if int(test) succeeded
        print(f'  ELSE   : Conversion succeeded! value = {value}')
    print()
````

## 13. `finally` — Always Runs

The `finally` block runs **no matter what** — whether an exception was raised or not, whether it was caught or not.

```
try → exception raised   → except runs  → finally RUNS
try → no exception       → else runs    → finally RUNS
```

**Common use cases for `finally`:**
- Closing a file or database connection
- Releasing a lock
- Printing a goodbye/cleanup message
- Any teardown that **must** happen regardless of outcome

````python
for test in ['5', 'bad']:
    print(f'--- Input: {test!r} ---')
    try:
        value = int(test)
    except ValueError:
        print('  EXCEPT  : Not a valid number')
    else:
        print(f'  ELSE    : Got value = {value}')
    finally:
        print('  FINALLY : This always runs (cleanup goes here)')
    print()
````

## 14. The Full `try / except / else / finally` Block

Here is the complete structure with all four clauses, and a visual summary of when each part runs:

```
┌──────────────────────────────────────────────────┐
│                   try block                      │
│         (code that might raise an error)         │
└────────────────────┬─────────────────────────────┘
                     │
          Exception? ├──── YES ──► except block
                     │
                    NO
                     │
                     ▼
                 else block
                 (success only)
                     │
                     ▼  (always, no matter what)
               finally block
```

````python
def process_input(raw):
    print(f'Processing {raw!r}...')
    try:
        number = int(raw)
        result = 100 / number
    except ValueError:
        print('  ✗ EXCEPT [ValueError]: Input is not a number')
    except ZeroDivisionError:
        print('  ✗ EXCEPT [ZeroDivisionError]: Cannot divide by zero')
    else:
        print(f'  ✓ ELSE: 100 / {number} = {result:.2f}')
    finally:
        print('  ■ FINALLY: Done processing\n')

process_input('4')       # success path
process_input('0')       # ZeroDivisionError
process_input('hello')   # ValueError
````

## 15. Common Built-in Exceptions

| Exception | When it happens | Example |
|---|---|---|
| `ValueError` | Right type, wrong value | `int('hello')` |
| `TypeError` | Wrong type entirely | `'2' + 2` |
| `ZeroDivisionError` | Division by zero | `1 / 0` |
| `IndexError` | List index out of range | `[1,2,3][10]` |
| `KeyError` | Dict key not found | `{'a':1}['b']` |
| `FileNotFoundError` | Opening a missing file | `open('nope.txt','r')` |
| `AttributeError` | Object has no such attribute | `'hi'.push(1)` |
| `NameError` | Variable not defined | `print(undefined_var)` |
| `PermissionError` | No access to file/resource | Writing to a read-only file |
| `OverflowError` | Number too large | `math.exp(9999)` |
| `RecursionError` | Too many recursive calls | Infinite recursion |
| `StopIteration` | Iterator is exhausted | Calling `next()` too many times |

````python
# Trigger and catch each common exception
import math

scenarios = [
    ('ValueError',         lambda: int('hello')),
    ('TypeError',          lambda: '2' + 2),
    ('ZeroDivisionError',  lambda: 1 / 0),
    ('IndexError',         lambda: [1, 2, 3][10]),
    ('KeyError',           lambda: {'a': 1}['b']),
    ('FileNotFoundError',  lambda: open('ghost.txt', 'r')),
    ('AttributeError',     lambda: 'hello'.nonexistent()),
    ('NameError',          lambda: eval('undefined_variable')),
]

for name, fn in scenarios:
    try:
        fn()
    except Exception as e:
        print(f'{type(e).__name__:22} → {e}')
````

## 16. Raising Exceptions — `raise`

You can **manually trigger** an exception using `raise`. This is useful when:
- Input fails your own validation rules
- You want to signal an error condition to the caller
- You want to re-raise a caught exception after logging it

```python
raise ValueError('Your custom message here')
raise TypeError('Expected a string')
raise   # re-raises the current active exception
```

````python
def set_age(age):
    if not isinstance(age, int):
        raise TypeError(f'Age must be an integer, got {type(age).__name__}')
    if age < 0 or age > 150:
        raise ValueError(f'Age {age} is out of realistic range (0–150)')
    return f'Age set to {age}'

test_cases = [25, -1, 200, 3.5, 'thirty']

for val in test_cases:
    try:
        print(set_age(val))
    except (ValueError, TypeError) as e:
        print(f'{type(e).__name__}: {e}')
````

## 17. Files + Exceptions Together

File operations are a prime case for exception handling — files can be missing, locked, or corrupted. Combining both skills produces robust, production-ready file code.

The most common file-related exceptions:

| Exception | Cause |
|---|---|
| `FileNotFoundError` | Trying to read a file that doesn't exist |
| `PermissionError` | No read/write access to the file |
| `IsADirectoryError` | Target path is a directory, not a file |
| `OSError` | General OS-level file error (parent of the above) |

````python
# Safe file reader — handles missing and unreadable files
def read_file_safely(filename):
    try:
        with open(filename, 'r') as f:
            content = f.read()
    except FileNotFoundError:
        print(f'  ✗ File not found: {filename!r}')
        return None
    except PermissionError:
        print(f'  ✗ Permission denied: {filename!r}')
        return None
    except OSError as e:
        print(f'  ✗ OS error: {e}')
        return None
    else:
        print(f'  ✓ Read {len(content)} characters from {filename!r}')
        return content
    finally:
        print(f'  ■ Attempted to read: {filename!r}\n')

# Test with an existing file and a missing file
read_file_safely('my_file.txt')   # exists
read_file_safely('ghost.txt')     # does not exist
````

````python
# Safe file writer with validation
def write_score(filename, name, score):
    if not isinstance(score, (int, float)):
        raise TypeError(f'Score must be a number, got {type(score).__name__}')
    if not (0 <= score <= 100):
        raise ValueError(f'Score {score} is out of range (0–100)')
    
    try:
        with open(filename, 'a') as f:
            f.write(f'{name}: {score}\n')
    except OSError as e:
        print(f'Failed to write: {e}')
    else:
        print(f'Saved: {name} → {score}')

# Valid entries
write_score('results.txt', 'Alice', 95)
write_score('results.txt', 'Bob',   72)

# Invalid entries — caught before writing
try:
    write_score('results.txt', 'Charlie', 110)   # out of range
except ValueError as e:
    print(f'ValueError: {e}')

try:
    write_score('results.txt', 'Diana', 'A+')    # wrong type
except TypeError as e:
    print(f'TypeError: {e}')

# Read back what was saved
print('\n--- results.txt ---')
with open('results.txt', 'r') as f:
    print(f.read())
````

## ✅ Chapter Summary

### 🗂️ Files

| Concept | Key Takeaway |
|---|---|
| `open(file, mode)` | Connects your program to a file on disk |
| `'r'` / `'w'` / `'a'` | Read / Write (overwrite) / Append |
| `'w'` mode | **Immediately erases** existing content — use with care |
| `write(str)` | Writes text — you must add `\n` manually |

| `close()` | Flushes data and releases the file — always required |
| `with open(...) as f:` | ✅ Best practice — file closes automatically |
| `for line in file:` | Memory-efficient way to read large files |

### ⚠️ Exceptions

| Concept | Key Takeaway |
|---|---|
| `try` | Code that might raise an exception |
| `except ExType as e` | Catches a specific exception; `e` holds the message |
| Multiple `except` | Handle different errors differently; specific before generic |
| `else` | Runs **only** on success (no exception in `try`) |
| `finally` | **Always** runs — ideal for cleanup/teardown |
| `raise ExType('msg')` | Manually trigger an exception with a custom message |
| `FileNotFoundError` | Most common file exception — always handle it |

### 🔗 Combined Pattern
```python
try:
    with open('file.txt', 'r') as f:
        data = f.read()
except FileNotFoundError:
    print('File not found!')
else:
    print('Success:', data)
finally:
    print('Done.')
```

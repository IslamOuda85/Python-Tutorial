## 1. String Testing Methods

These methods return `True` or `False` after inspecting the **contents** of a string.

| Method | Returns `True` when... |
|---|---|
| `isalnum()` | String contains **only** letters and/or digits (no special chars like `!@#`) |
| `isalpha()` | String contains **only** alphabetic letters |
| `isdigit()` | String contains **only** numeric digits |
| `isdecimal()` | String contains **only** decimal digits (same as `isdigit()` for simple cases) |
| `isnumeric()` | String contains **only** numeric characters |
| `islower()` | All alphabetic letters are lowercase (at least one letter must exist) |
| `isupper()` | All alphabetic letters are uppercase (at least one letter must exist) |
| `isspace()` | String contains **only** whitespace (spaces `' '`, tabs `'\t'`, newlines `'\n'`) |
| `istitle()` | First character of **each word** is uppercase, rest are lowercase |
| `isascii()` | All characters belong to the ASCII table |
| `isidentifier()` | String is a valid Python identifier (variable/keyword name) |
| `isprintable()` | All characters are printable (no invisible control characters) |

> ⚠️ All these methods require the string to have **at least one character** — an empty string returns `False`.

````python
# --- isalnum(), isalpha(), isdigit() ---

var2 = 'asd'
var3 = '123asd'
var4 = '123'
varx = '!@#'

print('--- isalnum() ---')
print('"asd"    isalnum():', var2.isalnum())   # True  (letters only)
print('"123asd" isalnum():', var3.isalnum())   # True  (letters + digits)
print('"123"    isalnum():', var4.isalnum())   # True  (digits only)
print('"!@#"    isalnum():', varx.isalnum())   # False (special chars)

print('\n--- isalpha() ---')
print('"asd"    isalpha():', var2.isalpha())   # True
print('"123asd" isalpha():', var3.isalpha())   # False (contains digits)

print('\n--- isdigit() ---')
print('"123"    isdigit():', var4.isdigit())   # True
print('"123asd" isdigit():', var3.isdigit())   # False
print('"asd"    isdigit():', var2.isdigit())   # False
````

````python
# --- islower(), isupper(), istitle() ---

print('--- islower() ---')
print('"ASD"      islower():', 'ASD'.islower())       # False
print('"asd"      islower():', 'asd'.islower())       # True
print('"asd qwe"  islower():', 'asd qwe'.islower())   # True  (spaces are ignored)
print('"Qaz"      islower():', 'Qaz'.islower())       # False (Q is uppercase)

print('\n--- isupper() ---')
print('"ISLAM OUDA" isupper():', 'ISLAM OUDA'.isupper())  # True
print('"islam ouda" isupper():', 'islam ouda'.isupper())  # False

print('\n--- istitle() ---')
print('"Islam Ouda" istitle():', 'Islam Ouda'.istitle())  # True
print('"islam ouda" istitle():', 'islam ouda'.istitle())  # False
print('"Islam ouda" istitle():', 'Islam ouda'.istitle())  # False (ouda not titled)
````

````python
# --- isspace(), isascii(), isidentifier(), isprintable() ---

print('--- isspace() ---')
print('" "       isspace():', ' '.isspace())          # True  (plain space)
print('"\\t"      isspace():', '\t'.isspace())         # True  (tab)
print('"\\n"      isspace():', '\n'.isspace())         # True  (newline)
print('"asd\\n"   isspace():', 'asd\n'.isspace())      # False (mixed content)
# raw string: the \n is treated as two characters, not a newline
print('r"\\t"     isspace():', r'\t'.isspace())        # False (raw: backslash + t)

print('\n--- isascii() ---')
print('"islam ouda" isascii():', 'islam ouda'.isascii())   # True
print('"íslåm øúðá" isascii():', 'íslåm øúðá'.isascii())  # False (non-ASCII)

print('\n--- isidentifier() ---')
print('"print"      isidentifier():', 'print'.isidentifier())       # True
print('"def"        isidentifier():', 'def'.isidentifier())         # True
print('"islam ouda" isidentifier():', 'islam ouda'.isidentifier())  # False (space)
print('"123var"     isidentifier():', '123var'.isidentifier())      # False (starts with digit)

print('\n--- isprintable() ---')
print('"islam ouda" isprintable():', 'islam ouda'.isprintable())  # True
print('"\\n"         isprintable():', '\n'.isprintable())          # False (control char)
````

## 2. String Modification Methods

These methods return a **new copy** of the string — strings in Python are immutable, so the original is never changed.

| Method | Description |
|---|---|
| `lower()` | Converts all letters to **lowercase** |
| `upper()` | Converts all letters to **uppercase** |
| `capitalize()` | Makes the **first character** uppercase, rest lowercase |
| `swapcase()` | Flips **every** character's case (upper↔lower) |
| `title()` | Makes the **first letter of each word** uppercase |
| `strip()` | Removes leading **and** trailing whitespace |
| `lstrip()` | Removes **leading** (left) whitespace or specified characters |
| `rstrip()` | Removes **trailing** (right) whitespace or specified characters |
| `replace(old, new)` | Replaces **all** occurrences of `old` with `new` |

> 💡 `lstrip(chars)` removes **any combination** of the given characters from the left — not a substring.

````python
# --- lower(), upper(), capitalize(), swapcase(), title() ---

var = 'IsLaM OuDa'

print('Original  :', var)
print('lower()   :', var.lower())       # 'islam ouda'
print('upper()   :', var.upper())       # 'ISLAM OUDA'
print('capitalize():', var.capitalize()) # 'Islam ouda' (first char upper, rest lower)
print('swapcase():', var.swapcase())    # 'iSlAm oUdA'
print('title()   :', var.title())       # 'Islam Ouda'
````

````python
# --- strip(), lstrip(), rstrip() ---

padded = '   hello world   '
print('strip()  :', repr(padded.strip()))   # 'hello world'
print('lstrip() :', repr(padded.lstrip()))  # 'hello world   '
print('rstrip() :', repr(padded.rstrip()))  # '   hello world'

# With a character argument — removes any matching chars from that end
var = 'islam ouda'
print('\nlstrip("i")   :', var.lstrip('i'))    # 'slam ouda'  (removes 'i' from left)
print('lstrip("is")  :', var.lstrip('is'))   # 'lam ouda'   (removes 'i' and 's')
print('lstrip("isl") :', var.lstrip('isl'))  # 'am ouda'    (removes 'i','s','l')
print('lstrip("qqq") :', var.lstrip('qqq'))  # 'islam ouda' (no match, unchanged)

# ⚠️ Important: lstrip removes individual chars, not a full substring
# 'isl' means: remove any 'i', 's', or 'l' found at the LEFT side
print('lstrip("sli") :', var.lstrip('sli'))  # same as lstrip('isl') = 'am ouda'
````

````python
# --- replace() ---

var = 'islam ouda'
print('replace("lam", "mail"):', var.replace('lam', 'mail'))  # 'ismail ouda'
print('replace("a", "@")     :', var.replace('a', '@'))       # 'isl@m oud@'  (ALL occurrences)

# replace() is case-sensitive
print('replace("Islam", "X") :', var.replace('Islam', 'X'))   # 'islam ouda'  (no change)
````

## 3. String Searching & Replacing Methods

| Method | Description |
|---|---|
| `find(sub)` | Returns the **lowest index** where `sub` is found; returns `-1` if not found |
| `index(sub)` | Exactly like `find()`, but raises a `ValueError` instead of returning `-1` |
| `count(sub)` | Returns the **number of times** `sub` appears in the string |
| `startswith(sub)` | Returns `True` if the string **starts** with `sub` |
| `endswith(sub)` | Returns `True` if the string **ends** with `sub` |

> 💡 **`find()` vs `index()`:** Use `find()` when you're not sure if the substring exists. Use `index()` when you *expect* it to be there and want an error if it's missing.

````python
var = 'islam ouda'
#       0123456789

# --- find() ---
print('--- find() ---')
print('find("m") :', var.find('m'))   # 4  → first 'm' is at index 4
print('find("a") :', var.find('a'))   # 3  → first 'a' (NOT 9)
print('find("e") :', var.find('e'))   # -1 → not found

# --- index() ---
print('\n--- index() ---')
print('index("") :', var.index(''))   # 0  → empty string found at position 0
try:
    var.index('z')                    # raises ValueError
except ValueError as e:
    print('index("z") raises ValueError:', e)

# --- count() ---
print('\n--- count() ---')
print('count("a") :', var.count('a'))  # 2  → 'a' appears twice
print('count("o") :', var.count('o'))  # 1
print('count("z") :', var.count('z'))  # 0  → no error, just 0

# --- startswith() & endswith() ---
print('\n--- startswith() & endswith() ---')
print('startswith("i") :', var.startswith('i'))   # True
print('startswith("a") :', var.startswith('a'))   # False
print('endswith("a")   :', var.endswith('a'))     # True
print('endswith("m")   :', var.endswith('m'))     # False
````

## 4. Alignment & Padding Methods

These methods are used to control how a string is positioned within a fixed-width field.

| Method | Description |
|---|---|
| `ljust(width)` | Left-aligns the string in a field of `width` characters (pads with spaces on the right) |
| `rjust(width)` | Right-aligns the string in a field of `width` characters (pads with spaces on the left) |
| `center(width)` | Centers the string within a field of `width` characters |
| `zfill(width)` | Pads the string with **zeros** on the left to reach the given width |

> ⚠️ If the given `width` is **smaller** than the string length, the string is returned unchanged — nothing is cut off.

````python
var = 'islam ouda'   # length = 10

# ljust: blocks 12 characters, string goes to the LEFT, spaces fill the RIGHT
print(repr(var.ljust(12)))    # 'islam ouda  '   (2 trailing spaces)
print(var.ljust(12), '***')   # islam ouda   ***

# rjust: blocks 15 characters, string goes to the RIGHT, spaces fill the LEFT
print(repr(var.rjust(15)))    # '     islam ouda'  (5 leading spaces)
print(var.rjust(15))          #      islam ouda

# width smaller than string length → no change
print(repr(var.ljust(5)))     # 'islam ouda'  (unchanged)

# center
print(repr(var.center(16)))   # '   islam ouda   '

# zfill: pads with zeros on the LEFT (commonly used for numbers stored as strings)
code = '42'
print(code.zfill(6))          # '000042'
print(var.zfill(14))          # '0000islam ouda'
````

## 5. Splitting & Partitioning Methods

| Method | Description |
|---|---|
| `split(sep)` | Splits the string on every occurrence of `sep`; returns a **list** |
| `splitlines(keepends)` | Splits on newlines only; `True` keeps `\n`, `False` removes it |
| `partition(sep)` | Splits at the **first** occurrence of `sep`; returns a **3-tuple**: `(before, sep, after)` |
| `rpartition(sep)` | Same as `partition()`, but searches from the **right** |

````python
# --- split() ---

var = 'islam ouda'
print('split()      :', var.split())         # ['islam', 'ouda']  (splits on whitespace)
print('split("a")   :', var.split('a'))      # ['isl', 'm oud', '']  (splits on every 'a')

sentence = 'one,two,three,four'
print('split(",")   :', sentence.split(',')) # ['one', 'two', 'three', 'four']

# --- splitlines() ---
var_nl = 'islam\nouda'
print('\nsplitlines(True) :', var_nl.splitlines(True))   # ['islam\n', 'ouda']  (keeps \n)
print('splitlines(False):', var_nl.splitlines(False))   # ['islam', 'ouda']   (strips \n)
print('split("\\n")     :', var_nl.split('\n'))          # same as splitlines(False)
````

````python
# --- partition() and rpartition() ---

var = 'islam ouda'

print('partition("l")  :', var.partition('l'))   # ('is', 'l', 'am ouda')  splits at FIRST 'l'
print('rpartition("a") :', var.rpartition('a'))  # ('islam oud', 'a', '')   splits at LAST 'a'

# Separator NOT found:
print('partition("z")  :', var.partition('z'))   # ('islam ouda', '', '')   whole string in [0]
print('rpartition("z") :', var.rpartition('z'))  # ('', '', 'islam ouda')  whole string in [2]
````

## 6. The `join()` Method

`join()` is the **opposite** of `split()`. It inserts the string (the one calling the method) **between** each character of the argument.

```python
separator.join(iterable)
```

| Argument type | Behaviour |
|---|---|
| A **list of strings** | Joins list items with the separator between them |
| A **single string** | Inserts separator between each **character** of the string |
| A single **1-char** string | Returns that one character unchanged |

> 💡 Common trick: use `''.join(reversed(s))` to **reverse a string** without slicing.

````python
var = 'islam ouda'

# join() on a string argument inserts the caller between each CHARACTER
print(var.join('{}'))     # {islam ouda}      → '{' + var + '}'
print(var.join(',,,'))    # ,islam ouda,islam ouda,   → var inserted between each ','
print(var.join('#.'))     # #islam ouda.       → var between '#' and '.'
print(var.join('.'))      # .                  → only 1 char → nothing to insert between

print()

# Most common use: joining a LIST of strings
words = ['my', 'name', 'is', 'islam']
print(' '.join(words))    # my name is islam
print('_'.join(words))    # my_name_is_islam
print(''.join(words))     # mynameislam

# Secret trick: reversing a string using join + reversed()
print(''.join(reversed(var)))  # aduo malsi
````

## 7. Prefix & Suffix Methods

These methods cleanly remove a specific **substring** from the start or end of a string.

| Method | Removes from... | Behaviour when not found |
|---|---|---|
| `removeprefix(prefix)` | The **beginning** | Returns the original string unchanged |
| `removesuffix(suffix)` | The **end** | Returns the original string unchanged |

> ⚠️ **Key difference from `lstrip`/`rstrip`:** These methods remove a **whole substring** — not individual characters.
> `lstrip('isl')` removes any `i`, `s`, or `l` chars from the left. `removeprefix('isl')` removes **exactly** the string `'isl'` if it's at the start.

````python
var = 'islam ouda'

print('removeprefix("isl") :', var.removeprefix('isl'))   # 'am ouda'
print('removeprefix("qqq") :', var.removeprefix('qqq'))   # 'islam ouda'  (not found → unchanged)

print('removesuffix("uda") :', var.removesuffix('uda'))   # 'islam o'
print('removesuffix("qqq") :', var.removesuffix('qqq'))   # 'islam ouda'  (not found → unchanged)

# Contrast with lstrip (removes individual chars, not a whole substring):
print('\nlstrip("isl")       :', var.lstrip('isl'))        # 'am ouda'  (same result here, but...)
print('lstrip("lis")       :', var.lstrip('lis'))         # 'am ouda'  (order doesn\'t matter for lstrip)
print('removeprefix("lis") :', var.removeprefix('lis'))   # 'islam ouda'  (must match exactly!)
````

## 8. Character Classification Methods

A quick reference for all the `is...()` methods grouped by what they test:

```
String content tested
├── Letters only       → isalpha()
├── Digits only        → isdigit()  /  isdecimal()  /  isnumeric()
├── Letters OR digits  → isalnum()
├── Whitespace only    → isspace()
├── All lowercase      → islower()
├── All uppercase      → isupper()
├── Title case         → istitle()
├── Valid identifier   → isidentifier()
├── ASCII range        → isascii()
└── All printable      → isprintable()
```

### `isdigit()` vs `isdecimal()` vs `isnumeric()`

For normal digit strings like `'123'`, all three behave the same. The difference appears with **special Unicode characters**:

| Method | Covers |
|---|---|
| `isdecimal()` | Standard decimal digits `0–9` only |
| `isdigit()` | Decimal digits + superscript digits (e.g. `²`) |
| `isnumeric()` | Above + fractions, Roman numerals, etc. |

````python
# For everyday use, all three give the same result:
var = '123'
print('isdecimal():', var.isdecimal())   # True
print('isdigit()  :', var.isdigit())     # True
print('isnumeric():', var.isnumeric())   # True

var = '123abc'
print('\nMixed string:')
print('isdecimal():', var.isdecimal())   # False
print('isdigit()  :', var.isdigit())     # False
print('isnumeric():', var.isnumeric())   # False

# Unicode superscript '²' — shows the difference:
superscript = '²'
print('\nSuperscript "²":')
print('isdecimal():', superscript.isdecimal())   # False
print('isdigit()  :', superscript.isdigit())     # True
print('isnumeric():', superscript.isnumeric())   # True
````

## 9. Miscellaneous Methods

| Method | Description |
|---|---|
| `encode(encoding, errors)` | Encodes the string into **bytes** using the specified encoding (e.g. `'utf-8'`) |
| `__hash__()` | Returns an integer **hash** of the string (strings are hashable and can be dict keys) |
| `__eq__(other)` | Returns `True` if the string's **value** equals `other`'s value |
| `translate(table)` | Replaces characters using a mapping table (advanced — covered after OOP) |

````python
# --- encode() ---
var = 'islam ouda'
print('encode(utf-8):', var.encode(encoding='utf-8', errors='strict'))  # b'islam ouda'
# The result is a bytes object (prefixed with b'')

# --- __hash__() and __eq__() ---
var  = 'islam ouda'
vare = 'islam ouda'
print('\nhash of var :', var.__hash__())     # some integer (random-looking, but consistent per session)
print('var == vare  :', var.__eq__(vare))   # True (same value)
print('var == "xyz" :', var.__eq__('xyz'))  # False

# Strings are hashable → they can be used as dictionary keys
my_dict = {var: 'found!'}
print('Used as key  :', my_dict['islam ouda'])
````

## 10. Text Formatting with `format()`

The `format()` method (and f-strings) lets you embed and style values inside a string using **format specifiers**.

### General Syntax inside `{}`
```
{index : fill align width .precision type}
   1       2    3     4       5        6
```

| Part | Meaning | Example |
|---|---|---|
| `index` | Which argument to use (0-based) | `{0}`, `{1}` |
| `fill` | Character to use for padding (default = space) | `#`, `0`, `*` |
| `align` | `<` left, `>` right, `^` center | `:<`, `:>`, `:^` |
| `width` | Minimum total field width (characters) | `:10` |
| `.precision` | Decimal places (for floats) | `:.2f` |
| `type` | `d` = integer, `f` = float, `s` = string | `:d`, `:f`, `:s` |

> 💡 The special alignment chars (`<`, `>`, `^`) go **between** the `:` and the width number.

````python
# --- Float precision ---
print("{:.2f}".format(4564.4533333))   # 4564.45  (2 decimal places)
print("{:.4f}".format(1532.754453434)) # 1532.7545 (4 decimal places, rounds up)

# round() does the same thing:
print(round(1532.754453434, 4))        # 1532.7545
````

````python
# --- Width (field blocking) ---
# {0:4} → argument 0 in a field 4 chars wide (right-aligned for numbers by default)
print("Sammy has {0:4} red {1:16}!".format(5, "balloons"))
#                   ^4^        ^----16----^
# Output: 'Sammy has    5 red balloons         !'
````

````python
# --- Alignment: <  >  ^ ---
#  <  =  LEFT    >  =  RIGHT    ^  =  CENTER

print("Left   |{:<20}|".format("I'm left"))          # text at left,  spaces right
print("Center |{:^20}|".format("I'm middle"))         # text centered
print("Right  |{:>20}|".format("I'm right"))          # text at right, spaces left
````

````python
# --- Custom fill character ---
# Put the fill char BEFORE the alignment symbol
print('{:#^16}'.format('islam'))    # ###islam######  (fills empty space with '#')
print('{:*<16}'.format('islam'))    # islam***********
print('{:0>16}'.format('42'))       # 00000000000000042
````

````python
# --- Combined: width + precision + type ---
# {0:5.0f} → argument 0, field width 5, 0 decimal places, float type
print("Sammy ate {0:5.0f} percent of a pizza!".format(75.765367))
# Output: 'Sammy ate    76 percent of a pizza!'  (rounds 75.76 → 76)

# Powers of 10 formatted as integers in fixed-width columns:
print('\nPowers of 10 (decimal format):')
for x in range(8):
    print(f'{10**x:10d}')   # each number right-aligned in 10-char field
````
